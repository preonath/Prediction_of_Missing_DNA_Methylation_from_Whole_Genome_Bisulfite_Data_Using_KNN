```
import pandas as pd
df = pd.read_csv('out.csv')
chr15_cgi = pd.read_csv('chr15_cgi.csv')
df['Loc_to_CGI'] = ""
for i in range(0, len(chr15_cgi)):
  df.loc[(df['start position'] >= chr15_cgi.iloc[i][1]) & (df['end position'] <= chr15_cgi.iloc[i][2]), 'Loc_to_CGI'] = 1
df.loc[df['Loc_to_CGI'] == '' , 'Loc_to_CGI'] = 0
df1 = df
df1.to_csv('Loc_to_CGI.csv')
Loc_to_CGI = pd.read_csv('Loc_to_CGI.csv')
Loc_to_CGI
```
