```
import pandas as pd
import pyranges as pr
cgi=pr.read_bed("CGI.bed",as_df=True)
cgi=cgi.loc[cgi["Chromosome"]=="chr15"]
cgi.to_csv("chr15_cgi.csv",index =False)
chr15_cgi=pd.read_csv("chr15_cgi.csv")
chr15_cgi
```

```
import pandas as pd
df = pd.read_csv('out.csv')
chr15_cgi = pd.read_csv('chr15_cgi.csv')
df['Loc to CGI'] = ""
for i in range(0, len(chr15_cgi)):
  df.loc[(df['start position'] >= chr15_cgi.iloc[i][1]) & (df['end position'] <= chr15_cgi.iloc[i][2]), 'Loc to CGI'] = True
df.loc[df['Loc to CGI'] == '' , 'Loc to CGI'] = False
df1 = df[['chromosome number','start position','end position','fraction of reads have dna methylation','Loc to CGI']]
df1.to_csv('Loc_to_CGI.csv')
Loc_to_CGI = pd.read_csv('Loc_to_CGI.csv')
Loc_to_CGI
```
